$\mu = \lambda * \Delta t \approx \Delta t / (\Delta x)^2$

```julia
f(μ) = (1-μ/2)/(1+μ/2) - exp(-μ)
g(μ) = (1-μ/2)/(1+μ/2)
```

```julia
x = -0.1:0.1:20
n = 4
plot(x,f.(x).^n,lw=2,ls=:dash)
plot!(x,g.(x).^n)
plot!(x,exp.(-x.*n))
```

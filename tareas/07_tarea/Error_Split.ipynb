Compute the error of the half/split method in QM

```julia
f(s) = (1 + s/2)/(1 - s/2) - exp(s)
```

```julia
x = 0.01:0.01:10
y = f.(1.0*im*x)
#y = f.(x)
```

```julia
using Plots
```

```julia
plot(x,abs.(y),xscale=:log10, yscale=:log10)
plot!(x,x.^3)
```

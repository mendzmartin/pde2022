# Estabilidad y error del método implícito

```julia
using LaTeXStrings;
```

Teniendo en cuenta que el método implícito, para nuestro problema particular, es de la forma,

\begin{align}
     \begin{array}{l}
    \{[ 2-\alpha _{1}( x_{i})]\hat{I} -\alpha \Delta t\hat{A}\} u_{1}( x_{i} ,t+\Delta t) -\beta u_{2}( x_{i} ,t+\Delta t) =\dotsc \\
    \dotsc \{[ 2+\alpha _{1}( x_{i})]\hat{I} +\alpha \Delta t\hat{A}\} u_{1}( x_{i} ,t) +\beta u_{2}( x_{i} ,t)
    \end{array}
\end{align}

y como los autovalores de $\hat{A}$ (operador que aproxima al Laplaceano $\Delta$) son de la forma $\lambda =\frac{-c}{( \Delta x)^{2}} ,\ c >0$ entonces para cada autofunción del operador $\hat{A}$ tendremos, de la ecuación implícita que,

\begin{align}
    [ 2-\alpha _{1}( x_{i}) -\alpha \Delta t\lambda _{k}] u_{1}^{k}( x_{i} ,t+\Delta t) =[ 2+\alpha _{1}( x_{i}) +\alpha \Delta t\lambda _{k}] u_{1}^{k}( x_{i} ,t) +\beta u_{2}^{k}( x_{i} ,t)
\end{align}

y el factor de amplificación para $u_1$ será (reemplazando valores para $\alpha ,\alpha _{i}( x) ,\beta $),

\begin{align}
    \begin{array}{l}
    f_{\mathrm{Amp}}\Bigl|_{u_{1}^{k}} := \left| \frac{[ 2-\alpha _{1}( x_{i}) -\alpha \Delta t\lambda _{k}]}{[ 2+\alpha _{1}( x_{i}) +\alpha \Delta t\lambda _{k}]}\right| =\left| \frac{\left[ 2+\frac{im\omega ^{2}( x-x_{a})^{2}}{2\hbar } -\frac{i\hbar \Delta t\lambda _{k}}{2m}\right]}{\left[ 2-\frac{im\omega ^{2}( x-x_{a})^{2}}{2\hbar } +\frac{i\hbar \Delta t\lambda _{k}}{2m}\right]}\right| \\
    \Rightarrow f_{\mathrm{Amp}}\Bigl|_{u_{1}^{k}} =\frac{\left| 4+i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{2\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right| }{\left| 4-i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{2\hbar } +\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right| } \equiv \frac{| s_{1} +is_{2}| }{| s_{1} -is_{2}| } ,( s_{i} \in \mathbb{R}) \Rightarrow \boxed{f_{\mathrm{Amp}}\Bigl|_{u_{1}^{k}} =1}\\
    \end{array}
\end{align}

y lo mismo ocurre para el factor de amplificación de $u_{2}^{k}$, esto es algo usual del método implícito para la ecuación de Schrödinger, donde el factor $a$ en la ecuación de la forma $\partial _{t} u( x,t) =a\Delta u( x,t)$ es imaginario puro. Entonces, el método implícito para estos casos es siempre estable, sin amplificar o disipar la solución al evolucionar el sistema, esto es muy importante para mantener la normalización de la función de onda.

Para analizar el error del método vemos que la solución general de la ecuación de Schrödinger, en caso particular en que el Hamiltoniano $\hat{H}$ es independiente del tiempo, será de la forma

\begin{align}
    \ket{\psi ( t)} =\sum _{k} A_{k}\exp\left( -\frac{iE_{k} t}{\hbar }\right)\ket{\psi _{E_{k}}}
\end{align}

donde los $A_n$ son coeficientes complejos y los estados $\ket{\psi _{E_{n}}}$ son soluciones a la ecuación estacionaria de Schrödinger (ec. de autovalores) de la forma $\hat{H}\ket{\psi _{E_{n}}} =E_{n}\ket{\psi _{E_{n}}}$. Con estos datos, sabemos el factor de amplificación del método y de la solución exacta y calculando su diferencia podremos saber el error del método,

\begin{align}
    \begin{array}{l}
    \epsilon _{\mathrm{imp}} =\left| f_{\mathrm{Amp}}^{\mathrm{exact}}\Bigl|_{u_{1}^{k}} -f_{\mathrm{Amp}}^{\mathrm{aprox}}\Bigl|_{u_{1}^{k}}\right| =\left| A_{k}\exp\left( -\frac{i\lambda _{k} t}{\hbar }\right) -\frac{( s_{1} +is_{2})}{( s_{1} -is_{2})}\right| ,\ ( s_{i} \in \mathbb{R})\\
    \Rightarrow \epsilon _{\mathrm{imp}} =\left| A_{k}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) -\frac{( s_{1} +is_{2}) is_{2}}{( s_{1})^{2}}\right| ,\ ( s_{i} \in \mathbb{R})
    \end{array}
\end{align}

donde se utilizó el desarrollo en serie de Taylor para el término $( s_{1} -is_{2})^{-1}$ a primer orden en $( \lambda _{k} \Delta t)$, al igual que en el desarrollo de la exponencial,

\begin{align}
    \exp\left( -\frac{i\lambda _{k} t}{\hbar }\right) =\sum _{n=0}^{\infty }\frac{\left(\frac{-i\lambda _{k} \Delta t}{\hbar }\right)^{n}}{n!} =\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) +\mathcal{O}\left(\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right)^{2}\right) \\

    \frac{1}{( 1-z)} =\sum _{n=0}^{\infty } z^{n} ,| z| < 1\Rightarrow \frac{1}{( s_{1} -is_{2})} =\frac{1}{s_{1}\left( 1-\frac{is_{2}}{s_{1}}\right)} =\frac{1}{s_{1}}\sum _{n=0}^{\infty }\left(\frac{is_{2}}{s_{1}}\right)^{n} ,\ ( s_{i} \in \mathbb{R}) \\

    \Rightarrow \frac{1}{( s_{1} -is_{2})} =\frac{is_{2}}{( s_{1})^{2}} +\mathcal{O}\left(\left(\frac{is_{2}}{( s_{1})^{2}}\right)^{2}\right) ,\ ( s_{i} \in \mathbb{R})
\end{align}

y reemplazando los valores para $s_i$,

\begin{align}
    \begin{array}{l}
    \epsilon _{\mathrm{imp}} =\left| A_{k}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) -\frac{\left\{4+i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right\} i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]}{( 4)^{2}}\right| \\
    \Rightarrow \epsilon _{\mathrm{imp}} =\left| A_{k}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) -\frac{1}{16}\left\{4+i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right\} i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right| 
    \end{array}
\end{align}

y para conocer simplemente el orden de magnitud del error tendremos,

\begin{align}
    \mathcal{O}( \epsilon _{\mathrm{imp}}) =\mathcal{O}\left(\left| A_{k}\left( -\frac{i}{\hbar } \lambda _{k} \Delta t\right) -\frac{1}{16}\left\{4+i\left[\frac{m\omega ^{2}( \Delta x)^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right\} i\left[\frac{m\omega ^{2}( \Delta x)^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right| \right)
\end{align}

y como $\mathcal{O}( \hbar ) =1$, $\mathcal{O}( m) =1$, $\mathcal{O}( \omega ) =1$, entonces (usando que $\lambda =\frac{-c}{( \Delta x)^{2}} ,\ c >0$)

\begin{align}
    \begin{array}{l}
    \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\frac{c\Delta t}{( \Delta x)^{2}} -\frac{1}{16}\left\{4+i\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]\right\}\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\frac{c\Delta t}{( \Delta x)^{2}} -\frac{1}{16}\left\{4\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right] +i\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]^{2}\right\}\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\frac{c\Delta t}{( \Delta x)^{2}} -\frac{1}{16}\left\{4( \Delta x)^{2} +\frac{4c\Delta t}{( \Delta x)^{2}} +i\left[( \Delta x)^{4} +2c\Delta t+\frac{( c\Delta t)^{2}}{( \Delta x)^{4}}\right]\right\}\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \left\{\left(\mathcal{O}( A_{k}) -\frac{1}{4}\right)\frac{c\Delta t}{( \Delta x)^{2}} -\frac{( \Delta x)^{2}}{4}\right\} -i\left\{\frac{\left[( \Delta x)^{4} +2c\Delta t+\frac{( c\Delta t)^{2}}{( \Delta x)^{4}}\right]}{16}\right\}\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \left(\mathcal{O}( A_{k}) -\frac{1}{4}\right)\frac{c\Delta t}{( \Delta x)^{2}} -\frac{( \Delta x)^{2}}{4}\right| +\frac{1}{16}\left[( \Delta x)^{4} +2c\Delta t+\frac{( c\Delta t)^{2}}{( \Delta x)^{4}}\right]\\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\frac{\left| 4\left(\mathcal{O}( A_{k}) -\frac{1}{4}\right) c\Delta t-( \Delta x)^{4}\right| }{4( \Delta x)^{2}} +\frac{\left[( \Delta x)^{16} +2c\Delta t( \Delta x)^{4} +( c\Delta t)^{2}\right]}{16( \Delta x)^{4}}\\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\frac{\left| ( 4\mathcal{O}( A_{k}) -1) c\Delta t-( \Delta x)^{4}\right| }{4( \Delta x)^{2}} +\frac{\left[( \Delta x)^{4} +c\Delta t\right]^{2}}{16( \Delta x)^{4}}\\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\frac{1}{4( \Delta x)^{2}}\left\{\left| ( 4\mathcal{O}( A_{k}) -1) c\Delta t-( \Delta x)^{4}\right| +\frac{\left[( \Delta x)^{4} +c\Delta t\right]^{2}}{4( \Delta x)^{2}}\right\}
    \end{array}
\end{align}

aquí el segundo término dentro de las llaves es el que nos da el orden de magnitud más grande para el error, entonces,

\begin{align}
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) \approx \frac{\left[( \Delta x)^{8} +2c( \Delta x)^{4} \Delta t+( \Delta t)^{2}\right]}{16( \Delta x)^{4}} \\

    \mathrm{si} \ \Delta x,\Delta t\ll 1\Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) \approx \left[( \Delta x)^{4} +\Delta t+\left(\frac{\Delta t}{( \Delta x)^{2}}\right)^{2}\right] \Rightarrow \boxed{\mathcal{O}( \epsilon _{\mathrm{imp}})\approx \left(\frac{\Delta t}{( \Delta x)^{2}}\right)^{2} \to \mathrm{a \space 1er \space orden \space}} 
\end{align}

donde hemos despreciado los primeros dos términos del corchete por sobre el tercero.

Si, en cambio, retenemos términos hasta segundo orden en $( \lambda _{k} \Delta t)$ para los desarrollos de Taylor tendremos,

\begin{align}
    \begin{array}{l}
    \exp\left( -\frac{i\lambda _{k} t}{\hbar }\right) =\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) +\frac{1}{2}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right)^{2} +\mathcal{O}\left(\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right)^{3}\right)\\
    \\
    \frac{1}{( s_{1} -is_{2})} =\frac{is_{2}}{( s_{1})^{2}} -\frac{( s_{2})^{2}}{( s_{1})^{3}} +\mathcal{O}\left(\left(\frac{is_{2}}{( s_{1})^{2}}\right)^{3}\right) ,\ ( s_{i} \in \mathbb{R})
    \end{array}
\end{align}

entonces,

\begin{align}
    \begin{array}{l}
    \epsilon _{\mathrm{imp}} =\left| A_{k}\left[\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) +\frac{1}{2}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right)^{2}\right] -\dotsc \right| \\
    \left| \dotsc -\frac{\left\{4+i\left[\tfrac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\tfrac{\hbar \Delta t\lambda _{k}}{m}\right]\right\}\left\{i\left[\tfrac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\tfrac{\hbar \Delta t\lambda _{k}}{m}\right] -\frac{1}{4}\left[\tfrac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\tfrac{\hbar \Delta t\lambda _{k}}{m}\right]^{2}\right\}}{( 4)^{2}}\right| 
    \end{array}
\end{align}

\begin{align}
    \begin{array}{l}
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\left[\left( i\frac{c\Delta t}{( \Delta x)^{2}}\right) -\frac{1}{2}\left(\frac{c\Delta t}{( \Delta x)^{2}}\right)^{2}\right] -\frac{\left\{4+i\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]\right\}\left\{i\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right] -\frac{1}{4}\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]^{2}\right\}}{16}\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\left[\left( i\frac{c\Delta t}{( \Delta x)^{2}}\right) -\frac{1}{2}\left(\frac{c\Delta t}{( \Delta x)^{2}}\right)^{2}\right] -\frac{\left\{-2\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]^{2} +\frac{i}{4}\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]\left\{16-\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]^{2}\right\}\right\}}{16}\right| 
    \end{array}
\end{align}

y los términos de mayor orden serán

\begin{align}
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) \approx \left[( \Delta x)^{8} +\left(\frac{\Delta t}{( \Delta x)^{2}}\right)^{3}\right] \Rightarrow \boxed{\mathcal{O}( \epsilon _{\mathrm{imp}}) \approx \left(\frac{\Delta t}{( \Delta x)^{2}}\right)^{3} \to \mathrm{a \space 2do \space orden \space}}
\end{align}

Entonces, como el método implícito es un método de 2do orden, tendremos que considerar el error reteniendo términos del desarrollo de Taylor hasta 2do orden, lo cual nos dice que el error escalea como el último resultado. Este último resultado fue el que hemos visto en el curso, sin embargo, el análisis anterior permite notar los ordenes que debemos retener en los desarrollos para reproducir dicho resultado.

Finalmente, notemos que el resultado anterior es válido para las regiones de convergencia de los desarrollos en serie de Taylor, es decir,

\begin{align}
    \begin{array}{l}
    \exp( x)\rightarrow x\ll 1\Rightarrow \left| \frac{-i\lambda _{k} \Delta t}{\hbar }\right| =\left| \frac{ic\Delta t}{\hbar ( \Delta x)^{2}}\right| \ll 1\Rightarrow \begin{cases}
    \hbar =1\\
    c=1
    \end{cases} \Rightarrow \boxed{\Delta t\ll ( \Delta x)^{2}}\\
    \frac{1}{( 1-z)}\rightarrow | z| < 1\Rightarrow \left| \frac{is_{2}}{s_{1}}\right| =\left| \frac{i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]}{4}\right| =\left| \frac{i\left[\frac{m\omega ^{2}( \Delta x)^{2}}{\hbar } +\frac{\hbar c\Delta t}{m( \Delta x)^{2}}\right]}{4}\right| < 1\\
    (\mathrm{a.u.}) \Rightarrow \begin{cases}
    \hbar =1\\
    c=1\\
    m=1\\
    \omega =1
    \end{cases} \Rightarrow \frac{( \Delta x)^{4} +\Delta t}{( \Delta x)^{2}} < 4\Rightarrow ( \Delta x)^{4} -4( \Delta x)^{2} +\Delta t< 0
    \end{array}
\end{align}

\begin{align}
    ( \Delta x)^{2} =\Lambda \Rightarrow \Lambda ^{2} -4\Lambda +\Delta t=0\Leftrightarrow \Lambda _{\pm } =( \Delta x)^{2} =\begin{cases}
    2+\sqrt{4-\Delta t}\\
    2-\sqrt{4-\Delta t}
    \end{cases} \Rightarrow \Delta x_{\pm \pm } =\begin{cases}
    2+\sqrt{4-\Delta t}\\
    2-\sqrt{4-\Delta t}\\
    -2-\sqrt{4-\Delta t}\\
    -2+\sqrt{4-\Delta t}
    \end{cases}
\end{align}

y tenemos cuatro valores posibles para $\Delta x$. Si proponemos que la discretización temporal utilizada en el método será tal que $0< \Delta t< 4\Rightarrow 0< \sqrt{4-\Delta t} < 2$, sin embargo, para que nos encontremos dentro de las regiones de convergencia deberemos cumplir que,

\begin{align}
    \begin{array}{l}
    \Rightarrow \left[( \Delta x)^{4} -4( \Delta x)^{2} +\Delta t\right] < 0\\
    \Rightarrow ( \Delta x-\Delta x_{++})( \Delta x-\Delta x_{+-})( \Delta x-\Delta x_{--})( \Delta x-\Delta x_{-+}) < 0
    \end{array}
\end{align}

\begin{align}
    \Rightarrow \left[ \Delta x-\underbrace{\left( 2+\sqrt{4-\Delta t}\right)}_{ >0}\right]\left[ \Delta x-\underbrace{\left( 2-\sqrt{4-\Delta t}\right)}_{ >0}\right]\left[ \Delta x-\underbrace{\left( -2-\sqrt{4-\Delta t}\right)}_{< 0}\right]\left[ \Delta x-\underbrace{\left( -2+\sqrt{4-\Delta t}\right)}_{< 0}\right] < 0
\end{align}

y como la discretización espacial siempre será tal que $\Delta x >0$ entonces deberá cumplirse que,

\begin{align}
    \boxed{[ 0< \Delta t< 4] \land \left[\left( 2-\sqrt{4-\Delta t}\right) < \Delta x< \left( 2+\sqrt{4-\Delta t}\right)\right]}
\end{align}

entonces, si por ejemplo, usamos que $\Delta x=0.1$, para cumplir con la primera condición tendremos que proponer que $\frac{\Delta t}{( \Delta x)^{2}} \ll 1$ y como regla general usaremos que $\Delta t=\alpha ( \Delta x)^{2} ;\ 0< \alpha \ll 1$ entonces, por la última condición que debemos cumplir tendremos,

\begin{align}
    \left( 2-\sqrt{4-\alpha ( \Delta x)^{2}}\right) < \Delta x< \left( 2+\sqrt{4-\alpha ( \Delta x)^{2}}\right) ;\ \alpha \ll 1
\end{align}

y si graficamos estas regiones tendremos,

![02_02_03_EstabilidadYErrorMetodoImplicito_01](./figuresPDE2022/02_02_03_EstabilidadYErrorMetodoImplicito_01.png)

entonces vemos que la condición $\Delta x >\left( 2-\sqrt{4-\alpha ( \Delta x)^{2}}\right)$ se cumple $\forall \alpha  >0$ y, a medida que disminuimos $\alpha$ obtenemos un rango de validez más amplio para $\Delta x$ posibles. Entonces, proponiendo la condición $\Delta t=\alpha ( \Delta x)^{2}$ con $0< \alpha \ll 1$ cumpliremos también la segunda condición, matemáticamente esto sería,

\begin{align}
    \begin{array}{l}
    \Delta x< \left( 2+\sqrt{4-\alpha ( \Delta x)^{2}}\right) \Rightarrow ( \Delta x-2)^{2} < \left| 4-\alpha ( \Delta x)^{2}\right| =4-\alpha ( \Delta x)^{2}\\
    \Rightarrow ( \Delta x)^{2} +4\Delta x+4-4+\alpha ( \Delta x)^{2} < 0\Rightarrow ( 1+\alpha )( \Delta x)^{2} +4\Delta x >0\\
    \Rightarrow \Delta x >-\frac{4}{( 1+\alpha )} \ \Rightarrow (\mathrm{se\ cumple} \ \forall \ \alpha  >0)\\
    \\
    \Delta x >\left( 2-\sqrt{4-\alpha ( \Delta x)^{2}}\right) \Rightarrow ( 2-\Delta x)^{2} < \left| 4-\alpha ( \Delta x)^{2}\right| =4-\alpha ( \Delta x)^{2}\\
    \Rightarrow ( \Delta x)^{2} -4\Delta x+4-4+\alpha ( \Delta x)^{2} < 0\Rightarrow ( 1+\alpha )( \Delta x)^{2} -4\Delta x< 0\\
    \Rightarrow \Delta x< \frac{4}{( 1+\alpha )}
    \end{array}
\end{align}

### 1. Declaramos parámetros constantes

```julia
const m=1.0;    # mass
const ω=1.0;    # angular frecuency
const ħ=1.0;    # Planck constant
const γ=0.1;    # coupling factor
const x₁=0.0;   # centering point QHO₁
const x₂=2.0;   # centering point QHO₂
```

### 2. Definimos algunas funciones útiles

```julia
λ(Δx)=-1.0/(Δx^2);                                 # como escalean los autovalores
s₂(Δx,Δt)=m*((ω*Δx)^2)/ħ-ħ*Δt*λ(Δx)*(1.0/m);   # factor imaginario
s₁=4.0;                                             # factor real
```

Construimos una función para calcular el factor de amplificación del método numérico

```julia
function f_AmpAprox(Δx_vect,Δt_vect;type="NOabs")
    dim_rows=length(Δx_vect);

    if type=="NOabs"
        MatrixResult=Matrix{ComplexF64}(undef,dim_rows,3);
    elseif type=="abs"
        MatrixResult=Matrix{Float64}(undef,dim_rows,3);
    end

    MatrixResult[:,1]=transpose(Δx_vect[:]);
    MatrixResult[:,2]=transpose(Δt_vect[:]);

    if type=="NOabs"
        for i in 1:dim_rows
            MatrixResult[i,3]=(s₁+im*s₂(Δx_vect[i],Δt_vect[i]))*(1.0/(s₁-im*s₂(Δx_vect[i],Δt_vect[i])));
        end
    elseif type=="abs"
        for i in 1:dim_rows
            # MatrixResult[i,3]=abs((s₁+im*s₂(Δx_vect[i],Δt_vect[i]))*(1.0/(s₁-im*s₂(Δx_vect[i],Δt_vect[i]))));
            MatrixResult[i,3]=abs((s₁+im*s₂(Δx_vect[i],Δt_vect[i])))*(1.0/abs(s₁-im*s₂(Δx_vect[i],Δt_vect[i])));
        end
    end

    return MatrixResult;
end
```

Construimos una función para calcular el factor de amplificación exácto

```julia
function f_AmpExact(Δx_vect,Δt_vect;type="NOabs")
    dim_rows=length(Δx_vect);

    if type=="NOabs"
        MatrixResult=Matrix{ComplexF64}(undef,dim_rows,3);
    elseif type=="abs"
        MatrixResult=Matrix{Float64}(undef,dim_rows,3);
    end

    MatrixResult[:,1]=transpose(Δx_vect[:]);
    MatrixResult[:,2]=transpose(Δt_vect[:]);

    if type=="NOabs"
        for i in 1:dim_rows
            MatrixResult[i,3]=exp(-im*λ(Δx_vect[i])*Δt_vect[i]*(1.0/ħ));
        end
    elseif type=="abs"
        for i in 1:dim_rows
            MatrixResult[i,3]=abs(exp(-im*λ(Δx_vect[i])*Δt_vect[i]*(1.0/ħ)));
        end
    end

    return MatrixResult;
end
```

Construimos una función para calcular el error del método implícito para problemas particulares de Mecánica Cuántica

```julia
function f_error(Δx_vect,Δt_vect;type="NOabs")
    dim_rows=length(Δx_vect);

    if type=="NOabs"
        MatrixResult=Matrix{ComplexF64}(undef,dim_rows,3);
    elseif type=="abs"
        MatrixResult=Matrix{Float64}(undef,dim_rows,3);
    end

    MatrixResult[:,1]=transpose(Δx_vect[:]);
    MatrixResult[:,2]=transpose(Δt_vect[:]);

    if type=="NOabs"
        for i in 1:dim_rows
            MatrixResult[i,3]=exp(-im*λ(Δx_vect[i])*Δt_vect[i]*(1.0/ħ))-(s₁+im*s₂(Δx_vect[i],Δt_vect[i]))*(1.0/(s₁-im*s₂(Δx_vect[i],Δt_vect[i])));
        end
    elseif type=="abs"
        for i in 1:dim_rows
            MatrixResult[i,3]=abs(exp(-im*λ(Δx_vect[i])*Δt_vect[i]*(1.0/ħ))-(s₁+im*s₂(Δx_vect[i],Δt_vect[i]))*(1.0/(s₁-im*s₂(Δx_vect[i],Δt_vect[i]))));
        end
    end
    
    return MatrixResult;
end
```

Ahora creamos una función para plotear el error aproximado (calculado haciendo desarrollos en serie de taylor hasta segundo orden en el factor $\frac{\Delta t}{\left( \Delta x \right)^2}$ que corresponde a cómo escalean los autovalores del operador de diferencias finitas en el método de elementos finitos.

```julia
function f_error_2ndOrder(Δx_vect,Δt_vect)
    dim_rows=length(Δx_vect);
    MatrixResult=Matrix{Float64}(undef,dim_rows,3);
    MatrixResult[:,1]=transpose(Δx_vect[:]);
    MatrixResult[:,2]=transpose(Δt_vect[:]);
    for i in 1:dim_rows
        MatrixResult[i,3]=(Δt_vect[i]/(Δx_vect[i]^2))^3
    end
    return MatrixResult;
end
```

```julia
Δx_2ndOrder=[0.01*i for i in 1:360];
Δt₁_2ndOrder=0.1.*Δx_2ndOrder;
f₁_2ndOrder=f_error_2ndOrder(Δx_2ndOrder,Δt₁_2ndOrder);

Δt₂_2ndOrder=0.01.*Δx_2ndOrder;
f₂_2ndOrder=f_error_2ndOrder(Δx_2ndOrder,Δt₂_2ndOrder);

plot1=scatter(f₁_2ndOrder[:,1],f₁_2ndOrder[:,3],label="\$\\Delta{t}=0.1\\Delta{x}\$",yscale=:log10)
plot1=scatter!(f₂_2ndOrder[:,1],f₂_2ndOrder[:,3],label="\$\\Delta{t}=0.01\\Delta{x}\$",yscale=:log10)
plot1=scatter!(xlabel="\$\\Delta x\$",ylabel="Error aprox. - \$\\left[\\Delta{t}/(\\Delta{x})^2\\right]^3\$")
savefig(plot1,path_plots*"ErrorAprox.pdf");
display(plot1)
```

## 1. Enunciado del problema

Sea un Hamiltoniano que describe la población de dos estados cuántico nucleares que poseen posiciones de equilibrio diferentes $x_a$ y $x_b$. Por otro lado, sea el Hamiltoniano nuclear uno de los estados,

\begin{align}
    \hat{h}_{x_{0}}( x) =-\frac{\hbar ^{2}}{2m}\frac{\partial ^{2}}{\partial x^{2}} +\frac{1}{2} m\omega ^{2}( x-x_{0})^{2}
\end{align}

entre los estados hay un término de interacción constatnte $\gamma$. De este modo podemos modelar el Hamiltoniano del sistema como,

\begin{align}
    \hat{H} =\begin{pmatrix}
    h_{x_{a}}( x) & \gamma \\
    \gamma  & h_{x_{a}}( x)
    \end{pmatrix} ;\ \gamma =\mathrm{cte}
\end{align}

y la ecuación de Schrödinger correspondiente es,

\begin{align}
    i\hbar \frac{\partial \Psi ( x,t)}{\partial t} =\hat{H} \Psi ( x,t) ;\ \Psi ( x,t) ;=\begin{pmatrix}
    \psi _{a}( x,t)\\
    \psi _{b}( x,t)
    \end{pmatrix}
\end{align}

Asumiendo que $\hbar=1$, $m=1$, $\omega=1$, $x_a=0$ y $x_b=2$. Sea también $\phi_n$ los autoestados de $\hat{h}_{x_{a} =0}$ y considere la condición inicial

\begin{align}
    \Psi _{0}( x,0) =\begin{pmatrix}
    0\\
    \phi _{0}( x,0)
    \end{pmatrix}
\end{align}

Es decir, el estado fundamental de $\hat{h}_{x_{a} =0}$ pero sobre el potencial de $\hat{h}_{x_{b} =0}$. COnsidere la interacción $\gamma=0.1$.

## 2. Primer método de resolución

Aquí la idea es resolver el problema d eautovalores para el Hamiltoniano unidimensional $\hat{h}_{x_{0}}(x)$ y luego construir el estado inicial del sistema. Después, para resolver la evolución temporal se utilizará el método de Euler implícito para discretizar la parte temporal del problema.

### 2.1 Problema de autovalores del hamiltoniano $\hat{h}_{x_{0}}$ unidimensional (resolución de autoestados en un espacio complejo)

En primera instancia buscamos diagonalizar el operador $\hat{h}_{x_{a}}(x)$ entonces queremos resolver la ec. de Schrödinger independiente del tiempo de la forma (ecuación de autovalores):

\begin{align}
    \hat{h}_{x_{a}}( x) \phi ( x) =\lambda \phi ( x) \Rightarrow \left[ -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2}}{\partial x^{2}} +\frac{1}{2} m\omega ^{2}( x-x_{a})^{2}\right] \phi ( x) =\lambda \phi ( x)
\end{align}

y según lo visto en el curso este problema es uno de autovalores, sin embargo, en el curso vimos un problema unidimensional de la forma

\begin{align}
    \mathrm{1D} \Rightarrow -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q( x) u( x) =\lambda r( x) u( x) ;\ p( x) ,r( x)  >0
\end{align}

y en nuestro problema particular tenemos una ecuación similar de Sturm-Liouville con parámetros $p(x)$,$q(x)$,$r(x)$ definidos como,

\begin{align}
    \begin{cases}
    p( x) =\frac{\hbar^{2}}{2m}\\
    q( x) =\left(\frac{m\omega ^{2}( x-x_{a})^{2}}{2} +\gamma \right)\\
    r( x) =1
    \end{cases} \Rightarrow \left[ -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q( x)\right] u( x) =\lambda r( x) u( x)
\end{align}

Las formas bilineales para considerar el problema débil de autovalores serán de la forma

\begin{align}
        a( u,v) &:= \int _{\Omega }[ p( x) \nabla v( x) \cdotp \nabla u( x) +q( x) v( x) u( x)] d\Omega \\
        b( u,v) &:= \int _{\Omega } r( x) u( x) v( x) d\Omega
\end{align}

aquí las funciones Test $v(x)$ y Trial $u(x)$ viven cada una en un espacio complejo.

### 2.2 Resolviendo la evolución temporal

Ahora queremos resolver la ecuación de Schrödinger dependiente del tiempo entonces, partiendo de lo visto en el curso, sabemos resolver problemas 1D de la forma

\begin{align}
    \frac{\partial \psi ( x,t)}{\partial t} =a\Delta \psi ( x,t) \Rightarrow \frac{\partial \psi ( x,t)}{\partial t} =a\frac{\partial ^{2} \psi ( x,t)}{\partial x^{2}}
\end{align}

en nuestro caso tendremos un sistema de ecuaciones de la forma

\begin{align}
     \begin{array}{l}
    i\hbar \frac{\partial \Psi ( x,t)}{\partial t} =\hat{H} \Psi ( x,t) \Rightarrow \frac{\partial }{\partial t}\begin{pmatrix}
    \psi _{a}\\
    \psi _{b}
    \end{pmatrix} =\left( -\frac{i}{\hbar }\right)\begin{pmatrix}
    h_{x_{a}}( x) & \gamma \\
    \gamma  & h_{x_{b}}( x)
    \end{pmatrix}\begin{pmatrix}
    \psi _{a}( x,t)\\
    \psi _{b}( x,t)
    \end{pmatrix}\\
    \Rightarrow \begin{cases}
    \frac{\partial \psi _{a}( x,t)}{\partial t} =\left(\frac{-i}{\hbar }\right)[ h_{x_{a}}( x) \psi _{a}( x,t) +\gamma \psi _{b}( x,t)]\\
    \frac{\partial \psi _{b}( x,t)}{\partial t} =\left(\frac{-i}{\hbar }\right)[ \gamma \psi _{a}( x,t) +h_{x_{b}}( x) \psi _{b}( x,t)]
    \end{cases}\\
    \Rightarrow \begin{cases}
    \frac{\partial \psi _{a}( x,t)}{\partial t} =\left(\frac{-i}{\hbar }\right)\left\{-\frac{\hbar ^{2}}{2m}\frac{\partial ^{2} \psi _{a}( x,t)}{\partial x^{2}} +\left[\frac{1}{2} m\omega ^{2}( x-x_{a})^{2} \psi _{a}( x,t) +\gamma \psi _{b}( x,t)\right]\right\}\\
    \frac{\partial \psi _{b}( x,t)}{\partial t} =\left(\frac{-i}{\hbar }\right)\left\{-\frac{\hbar ^{2}}{2m}\frac{\partial ^{2} \psi _{b}( x,t)}{\partial x^{2}} +\left[\frac{1}{2} m\omega ^{2}( x-x_{b})^{2} \psi _{b}( x,t) +\gamma \psi _{a}( x,t)\right]\right\}
    \end{cases}
    \end{array}
\end{align}

Para la ecuación 1D sabemos que el Laplaceano lo podremos aproximar en diferencias finitas como un operador $\hat{A}$ entonces el problema de forma discreta (espacialmente) se podrá expresar como

\begin{align}
    \frac{\partial u( x_{i} ,t)}{\partial t} =a\hat{A} u( x_{i} ,t)
\end{align}

donde usamos $u( x_{i} ,t)$ para referirnos a la función aproximada (discretizada espacialmente) asociada a $\psi(x,t)$. Y ahora discretizando la parte temporal del problema mediante el método de Euler Implícito, expresando la derivada temporal como una diferenciación hacia adelante y expresando la función $u( x_{i} ,t)$ como un promedio entre los pasos temporales actual y posterior tendremos,

\begin{align}
     \begin{array}{l}
    \frac{[ u( x_{i} ,( t_{j} +\Delta t)) -u( x_{i} ,t_{j})]}{\Delta t} =a\hat{A}\frac{[ u( x_{i} ,t_{j}) +u( x_{i} ,( t_{j} +\Delta t))]}{2}\\
    \Rightarrow ( 2\hat{I} -a\Delta t\hat{A}) u( x_{i} ,( t_{j} +\Delta t)) =( 2\hat{I} +a\Delta t\hat{A}) u( x_{i} ,t_{j})
    \end{array}
\end{align}

y las formas bilineales asociadas a este problema serán

\begin{align}
    a( u,v) &:= \int _{\Omega }\{2u( x_{i} ,t_{j} +\Delta t) v( x_{i} ,t_{j} +\Delta t) -a[ \nabla v( x_{i} ,t_{j} +\Delta t) \cdotp \nabla u( x_{i} ,t_{j} +\Delta t)] \Delta t\} d\Omega \\
    b( v) &:= \int _{\Omega }\{2u( x_{i} ,t_{j}) v( x_{i} ,t_{j}) +a[ \nabla v( x_{i} ,t_{j}) \cdotp \nabla u( x_{i} ,t_{j})] \Delta t\} d\Omega
\end{align}

Y este problema débil deberá resolverse para cada tiempo, lo cual implicará que el funcional $b(v)$ deba estar dentro de un bucle `for` barriendo los $t_{j} =( t_{o} +j\cdotp \Delta t)$ con $j\in \mathbb{N}$.

Ahora bien, para nuestro caso tenemos el siguiente sistema de ecuaciones a resolver,

\begin{align}
    \begin{cases}
    \frac{\partial \psi _{a}( x,t)}{\partial t} =\alpha \frac{\partial ^{2} \psi _{a}( x,t)}{\partial x^{2}} +\alpha _{1}( x) \psi _{a}( x,t) +\beta \psi _{b}( x,t)\\
    \frac{\partial \psi _{b}( x,t)}{\partial t} =\alpha \frac{\partial ^{2} \psi _{b}( x,t)}{\partial x^{2}} +\alpha _{2}( x) \psi _{b}( x,t) +\beta \psi _{a}( x,t)
    \end{cases}
\end{align}

donde hemos definido $\alpha := \frac{i\hbar }{2m} ;\alpha _{i}( x) := \frac{-im\omega ^{2}( x-x_{i})^{2}}{2\hbar } ;\beta := \frac{-i\gamma }{\hbar }$. Aproximando el operador Laplaceano por un operador de diferencias finitas podremos discretizar la parte espacial d ela forma,

\begin{align}
    \begin{cases}
    \frac{\partial u_{1}( x_{i} ,t)}{\partial t} =\alpha \hat{A} u_{1}( x_{i} ,t) +\alpha _{1}( x_{i}) u_{1}( x_{i} ,t) +\beta u_{2}( x_{i} ,t)\\
    \frac{\partial u_{2}( x_{i} ,t)}{\partial t} =\alpha \hat{A} u_{2}( x_{i} ,t) +\alpha _{2}( x_{i}) u_{2}( x_{i} ,t) +\beta u_{1}( x_{i} ,t)
    \end{cases}
\end{align}

Ahora, aplicando el método de Euler implícito para discretizar la parte temporal (de forma análoga al caso anterior). Entonces, comenzando con la primera ecuación tendremos,

\begin{align}
     \begin{array}{l}
    \frac{[ u_{1}( x_{i} ,t_{j} +\Delta t) -u_{1}( x_{i} ,t_{j})]}{\Delta t} =\alpha \hat{A}\frac{[ u_{1}( x_{i} ,t_{j}) +u_{1}( x_{i} ,t_{j} +\Delta t)]}{2} +\dotsc \\
    \dotsc +\alpha _{1}( x_{i})\frac{[ u_{1}( x_{i} ,t_{j}) +u_{1}( x_{i} ,t_{j} +\Delta t)]}{2} +\beta \frac{[ u_{2}( x_{i} ,t_{j}) +u_{2}( x_{i} ,t_{j} +\Delta t)]}{2}\\
    \Rightarrow \{2\hat{I} -[ \alpha \hat{A} +\alpha _{1}( x_{i})] \Delta t\} u_{1}( x_{i} ,t_{j} +\Delta t) -\beta \Delta t u_{2}( x_{i} ,t_{j} +\Delta t) =\dotsc \\
    \dotsc \{2\hat{I} +[ \alpha \hat{A} +\alpha _{1}( x_{i})] \Delta t\} u_{1}( x_{i} ,t_{j}) +\beta \Delta tu_{2}( x_{i} ,t_{j})
    \end{array}
\end{align}

y de forma análoga para la segunda ecuación tendremos,

\begin{align}
     \begin{array}{l}
    \Rightarrow \{2\hat{I} -[ \alpha \hat{A} +\alpha _{2}( x_{i})] \Delta t\} u_{2}( x_{i} ,t_{j} +\Delta t) -\beta \Delta t u_{2}( x_{i} ,t_{j} +\Delta t) =\dotsc \\
    \dotsc \{2\hat{I} +[ \alpha \hat{A} +\alpha _{2}( x_{i})] \Delta t\} u_{2}( x_{i} ,t_{j}) +\beta \Delta tu_{1}( x_{i} ,t_{j})
    \end{array}
\end{align}

entonces tendremos el siguiente sistema de ecuaciones,

\begin{align}
     \begin{array}{l}
    \begin{pmatrix}
    \{2\hat{I} -[ \alpha \hat{A} +\alpha _{1}( x_{i})] \Delta t\} & -\beta \Delta t\\
    -\beta \Delta t & \{2\hat{I} -[ \alpha \hat{A} +\alpha _{2}( x_{i})] \Delta t\}
    \end{pmatrix}\begin{pmatrix}
    u_{1}( x_{i} ,t+\Delta t)\\
    u_{2}( x_{i} ,t+\Delta t)
    \end{pmatrix} =\\
    \begin{pmatrix}
    \{2\hat{I} +[ \alpha \hat{A} +\alpha _{1}( x_{i})] \Delta t\} & \beta \Delta t\\
    \beta \Delta t & \{2\hat{I} +[ \alpha \hat{A} +\alpha _{2}( x_{i})] \Delta t\}
    \end{pmatrix}\begin{pmatrix}
    u_{1}( x_{i} ,t)\\
    u_{2}( x_{i} ,t)
    \end{pmatrix}
    \end{array}
\end{align}

Y las formas tensoriales débiles para este caso serán

\begin{align}
    \begin{array}{l}
    a_{1}(( u_{1} ,u_{2}) ,v_{1}) := \int _{\Omega }\{2( u_{1} v_{1}) -[ -\alpha ( \nabla v_{1} \cdotp \nabla u_{1}) +\alpha _{1}( u_{1} v_{1}) +\beta ( u_{2} v_{1})] \Delta t\} d\Omega \\
    a_{2}(( u_{2} ,u_{1}) ,v_{2}) := \int _{\Omega }\{2( u_{2} v_{2}) -[ -\alpha ( \nabla v_{2} \cdotp \nabla u_{2}) +\alpha _{2}( u_{2} v_{2}) +\beta ( u_{1} v_{2})] \Delta t\} d\Omega \\
    \\
    \Rightarrow a(( u_{1} ,v_{1}) ,( u_{2} ,v_{2})) := a_{1}(( u_{1} ,u_{2}) ,v_{1}) +a_{2}(( u_{2} ,u_{1}) ,v_{2})\\
    \\
    \\
    b_{1}( v_{1}) := \int _{\Omega }\left\{2\left( u_{1}^{0} v_{1}\right) +\left[ -\alpha \left( \nabla v_{1} \cdotp \nabla u_{1}^{0}\right) +\alpha _{1}\left( u_{1}^{0} v_{1}\right) +\beta \left( u_{2}^{0} v_{1}\right)\right] \Delta t\right\} d\Omega \\
    b_{2}( v_{2}) := \int _{\Omega }\left\{2\left( u_{2}^{0} v_{2}\right) +\left[ -\alpha \left( \nabla v_{2} \cdotp \nabla u_{2}^{0}\right) +\alpha _{2}\left( u_{2}^{0} v_{2}\right) +\beta \left( u_{1}^{0} v_{2}\right)\right] \Delta t\right\} d\Omega \\
    \\
    \Rightarrow b( v_{1} ,v_{2}) := b_{1}( v_{1}) +b_{2}( v_{2})
    \end{array}
\end{align}

y resolveremos una ecuación del tipo

\begin{align}
    a(( u_{1} ,v_{1}) ,( u_{2} ,v_{2}))\vec{u}( x_{i} ,t_{j}) =b( v_{1} ,v_{2}) \Rightarrow \boxed{\vec{u}( x_{i} ,t_{j}) =\left( a^{-1}\right) b}
\end{align}

y tendremos que resolver un problema de inversión de matrices. Aquí, la forma bilineal $b(v_1,v_2)$ deberá actualizarse en cada paso evolucionado con un nuevo par inicial $\vec{u}^{0} =\left( u_{1}^{0} ,u_{2}^{0}\right)$.

#### 2.2.1. Construyendo el operador densidad

Como resultado de la implementación numérica tendremos una función vectorial aproximada de la forma,

\begin{align}
    \vec{u} =\begin{pmatrix}
    u_{1}\\
    u_{2}
    \end{pmatrix} =\begin{pmatrix}
    [\mathrm{\Re e}( u_{1}) +i\mathrm{\Im m}( u_{1})]\\
    [\mathrm{\Re e}( u_{2}) +i\mathrm{\Im m}( u_{2})]
    \end{pmatrix}
\end{align}

y en Visit (o dentro de Julia mismo) podremos calcular la densidad de probabilidad como

\begin{align}
    \hat{\rho }( x_{i} ,t_{j}) := [\vec{u}( x_{i} ,t_{j})]^{\dagger } \cdotp \vec{u}( x_{i} ,t_{j})
\end{align}

y definiendo $u_{1,2}^{\Re }( x_{i} ,t_{j}) := \mathrm{\Re e}( u_{1,2}( x_{i} ,t_{j}))$, $u_{1,2}^{\mathrm{\Im }}( x_{i} ,t_{j}) := \mathrm{\Im m}( u_{1,2}( x_{i} ,t_{j}))$ tendremos,

\begin{align}
    \Rightarrow \hat{\rho }( x_{i} ,t_{j}) =\begin{pmatrix}
    \left[ u_{1}^{\Re }( x_{i} ,t_{j}) -i\mathrm{\Im m}( u_{1}( x_{i} ,t_{j}))\right] & \left[ u_{2}^{\Re }( x_{i} ,t_{j}) -i\mathrm{\Im m}( u_{2}( x_{i} ,t_{j}))\right]
    \end{pmatrix} \cdotp \begin{pmatrix}
    \left[ u_{1}^{\Re }( x_{i} ,t_{j}) +i\mathrm{\Im m}( u_{1}( x_{i} ,t_{j}))\right]\\
    \left[ u_{2}^{\Re }( x_{i} ,t_{j}) +i\mathrm{\Im m}( u_{2}( x_{i} ,t_{j}))\right]
    \end{pmatrix}
\end{align}

\begin{align}
    \begin{array}{l}
    \Rightarrow \hat{\rho }( x_{i} ,t_{j}) =\left[ u_{1}^{\Re }( x_{i} ,t_{j}) -iu_{1}^{\mathrm{\Im }}( x_{i} ,t_{j})\right]\left[ u_{1}^{\Re }( x_{i} ,t_{j}) +iu_{1}^{\mathrm{\Im }}( x_{i} ,t_{j})\right] +\dotsc \\
    \dotsc +\left[ u_{2}^{\Re }( x_{i} ,t_{j}) -iu_{2}^{\mathrm{\Im }}( x_{i} ,t_{j})\right]\left[ u_{2}^{\Re }( x_{i} ,t_{j}) +iu_{2}^{\mathrm{\Im }}( x_{i} ,t_{j})\right]\\
    \Rightarrow \hat{\rho }( x_{i} ,t_{j}) =\left[ u_{1}^{\Re }( x_{i} ,t_{j})\right]^{2} +\left[ u_{1}^{\mathrm{\Im }}( x_{i} ,t_{j})\right]^{2} +\left[ u_{2}^{\Re }( x_{i} ,t_{j})\right]^{2} +\left[ u_{2}^{\mathrm{\Im }}( x_{i} ,t_{j})\right]^{2}\\
    \Rightarrow \boxed{\hat{\rho }( x_{i} ,t_{j}) =\sum _{k=1}^{2}| u_{k}( x_{i} ,t_{j})| ^{2}}
    \end{array}
\end{align}

#### 2.2.2. Cálculo de poblaciones

Una vez obtenido el operador densidad podremos calcular las probabilidades de que una dada partícula esté en un volumen $V$ para un dado tiempo $t$ de la siguiente manera,

\begin{align}
    p( t) =\int _{V} \rho (\vec{x} ,t) dV
\end{align}

y numéricamente tendremos

\begin{align}
    p_{k}( t_{j}) =\mathrm{sum}\left(\int _{\Omega }[ u_{k}( x_{i} ,t_{j})]^{*} u_{k}( x_{i} ,t_{j}) d\Omega \right) ;\ k=\{1,2\}
\end{align}

#### 2.2.3. Estabilidad y error del método de Euler implícito

Teniendo en cuenta que el método implícito, para nuestro problema particular, es de la forma,

\begin{align}
     \begin{array}{l}
    \{[ 2-\alpha _{1}( x_{i})]\hat{I} -\alpha \Delta t\hat{A}\} u_{1}( x_{i} ,t+\Delta t) -\beta u_{2}( x_{i} ,t+\Delta t) =\dotsc \\
    \dotsc \{[ 2+\alpha _{1}( x_{i})]\hat{I} +\alpha \Delta t\hat{A}\} u_{1}( x_{i} ,t) +\beta u_{2}( x_{i} ,t)
    \end{array}
\end{align}

y como los autovalores de $\hat{A}$ (operador que aproxima al Laplaceano $\Delta$) son de la forma $\lambda =\frac{-c}{( \Delta x)^{2}} ,\ c >0$ entonces para cada autofunción del operador $\hat{A}$ tendremos, de la ecuación implícita que,

\begin{align}
    [ 2-\alpha _{1}( x_{i}) -\alpha \Delta t\lambda _{k}] u_{1}^{k}( x_{i} ,t+\Delta t) =[ 2+\alpha _{1}( x_{i}) +\alpha \Delta t\lambda _{k}] u_{1}^{k}( x_{i} ,t) +\beta u_{2}^{k}( x_{i} ,t)
\end{align}

y el factor de amplificación para $u_1$ será (reemplazando valores para $\alpha ,\alpha _{i}( x) ,\beta $),

\begin{align}
    \begin{array}{l}
    f_{\mathrm{Amp}}\Bigl|_{u_{1}^{k}} := \left| \frac{[ 2-\alpha _{1}( x_{i}) -\alpha \Delta t\lambda _{k}]}{[ 2+\alpha _{1}( x_{i}) +\alpha \Delta t\lambda _{k}]}\right| =\left| \frac{\left[ 2+\frac{im\omega ^{2}( x-x_{a})^{2}}{2\hbar } -\frac{i\hbar \Delta t\lambda _{k}}{2m}\right]}{\left[ 2-\frac{im\omega ^{2}( x-x_{a})^{2}}{2\hbar } +\frac{i\hbar \Delta t\lambda _{k}}{2m}\right]}\right| \\
    \Rightarrow f_{\mathrm{Amp}}\Bigl|_{u_{1}^{k}} =\frac{\left| 4+i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{2\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right| }{\left| 4-i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{2\hbar } +\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right| } \equiv \frac{| s_{1} +is_{2}| }{| s_{1} -is_{2}| } ,( s_{i} \in \mathbb{R}) \Rightarrow \boxed{f_{\mathrm{Amp}}\Bigl|_{u_{1}^{k}} =1}\\
    \end{array}
\end{align}

y lo mismo ocurre para el factor de amplificación de $u_{2}^{k}$, esto es algo usual del método implícito para la ecuación de Schrödinger, donde el factor $a$ en la ecuación de la forma $\partial _{t} u( x,t) =a\Delta u( x,t)$ es imaginario puro. Entonces, el método implícito para estos casos es siempre estable, sin amplificar o disipar la solución al evolucionar el sistema, esto es muy importante para mantener la normalización de la función de onda.

Para analizar el error del método vemos que la solución general de la ecuación de Schrödinger, en caso particular en que el Hamiltoniano $\hat{H}$ es independiente del tiempo, será de la forma

\begin{align}
    \ket{\psi ( t)} =\sum _{k} A_{k}\exp\left( -\frac{iE_{k} t}{\hbar }\right)\ket{\psi _{E_{k}}}
\end{align}

donde los $A_n$ son coeficientes complejos y los estados $\ket{\psi _{E_{n}}}$ son soluciones a la ecuación estacionaria de Schrödinger (ec. de autovalores) de la forma $\hat{H}\ket{\psi _{E_{n}}} =E_{n}\ket{\psi _{E_{n}}}$. Con estos datos, sabemos el factor de amplificación del método y de la solución exacta y calculando su diferencia podremos saber el error del método,

\begin{align}
    \begin{array}{l}
    \epsilon _{\mathrm{imp}} =\left| f_{\mathrm{Amp}}^{\mathrm{exact}}\Bigl|_{u_{1}^{k}} -f_{\mathrm{Amp}}^{\mathrm{aprox}}\Bigl|_{u_{1}^{k}}\right| =\left| A_{k}\exp\left( -\frac{i\lambda _{k} t}{\hbar }\right) -\frac{( s_{1} +is_{2})}{( s_{1} -is_{2})}\right| ,\ ( s_{i} \in \mathbb{R})\\
    \Rightarrow \epsilon _{\mathrm{imp}} =\left| A_{k}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) -\frac{( s_{1} +is_{2}) is_{2}}{( s_{1})^{2}}\right| ,\ ( s_{i} \in \mathbb{R})
    \end{array}
\end{align}

donde se utilizó el desarrollo en serie de Taylor para el término $( s_{1} -is_{2})^{-1}$ a primer orden en $( \lambda _{k} \Delta t)$, al igual que en el desarrollo de la exponencial,

\begin{align}
    \exp\left( -\frac{i\lambda _{k} t}{\hbar }\right) =\sum _{n=0}^{\infty }\frac{\left(\frac{-i\lambda _{k} \Delta t}{\hbar }\right)^{n}}{n!} =\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) +\mathcal{O}\left(\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right)^{2}\right) \\

    \frac{1}{( 1-z)} =\sum _{n=0}^{\infty } z^{n} ,| z| < 1\Rightarrow \frac{1}{( s_{1} -is_{2})} =\frac{1}{s_{1}\left( 1-\frac{is_{2}}{s_{1}}\right)} =\frac{1}{s_{1}}\sum _{n=0}^{\infty }\left(\frac{is_{2}}{s_{1}}\right)^{n} ,\ ( s_{i} \in \mathbb{R}) \\

    \Rightarrow \frac{1}{( s_{1} -is_{2})} =\frac{is_{2}}{( s_{1})^{2}} +\mathcal{O}\left(\left(\frac{is_{2}}{( s_{1})^{2}}\right)^{2}\right) ,\ ( s_{i} \in \mathbb{R})
\end{align}

y reemplazando los valores para $s_i$,

\begin{align}
    \begin{array}{l}
    \epsilon _{\mathrm{imp}} =\left| A_{k}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) -\frac{\left\{4+i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right\} i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]}{( 4)^{2}}\right| \\
    \Rightarrow \epsilon _{\mathrm{imp}} =\left| A_{k}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) -\frac{1}{16}\left\{4+i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right\} i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right| 
    \end{array}
\end{align}

y para conocer simplemente el orden de magnitud del error tendremos,

\begin{align}
    \mathcal{O}( \epsilon _{\mathrm{imp}}) =\mathcal{O}\left(\left| A_{k}\left( -\frac{i}{\hbar } \lambda _{k} \Delta t\right) -\frac{1}{16}\left\{4+i\left[\frac{m\omega ^{2}( \Delta x)^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right\} i\left[\frac{m\omega ^{2}( \Delta x)^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]\right| \right)
\end{align}

y como $\mathcal{O}( \hbar ) =1$, $\mathcal{O}( m) =1$, $\mathcal{O}( \omega ) =1$, entonces (usando que $\lambda =\frac{-c}{( \Delta x)^{2}} ,\ c >0$)

\begin{align}
    \begin{array}{l}
    \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\frac{c\Delta t}{( \Delta x)^{2}} -\frac{1}{16}\left\{4+i\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]\right\}\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\frac{c\Delta t}{( \Delta x)^{2}} -\frac{1}{16}\left\{4\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right] +i\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]^{2}\right\}\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\frac{c\Delta t}{( \Delta x)^{2}} -\frac{1}{16}\left\{4( \Delta x)^{2} +\frac{4c\Delta t}{( \Delta x)^{2}} +i\left[( \Delta x)^{4} +2c\Delta t+\frac{( c\Delta t)^{2}}{( \Delta x)^{4}}\right]\right\}\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \left\{\left(\mathcal{O}( A_{k}) -\frac{1}{4}\right)\frac{c\Delta t}{( \Delta x)^{2}} -\frac{( \Delta x)^{2}}{4}\right\} -i\left\{\frac{\left[( \Delta x)^{4} +2c\Delta t+\frac{( c\Delta t)^{2}}{( \Delta x)^{4}}\right]}{16}\right\}\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \left(\mathcal{O}( A_{k}) -\frac{1}{4}\right)\frac{c\Delta t}{( \Delta x)^{2}} -\frac{( \Delta x)^{2}}{4}\right| +\frac{1}{16}\left[( \Delta x)^{4} +2c\Delta t+\frac{( c\Delta t)^{2}}{( \Delta x)^{4}}\right]\\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\frac{\left| 4\left(\mathcal{O}( A_{k}) -\frac{1}{4}\right) c\Delta t-( \Delta x)^{4}\right| }{4( \Delta x)^{2}} +\frac{\left[( \Delta x)^{16} +2c\Delta t( \Delta x)^{4} +( c\Delta t)^{2}\right]}{16( \Delta x)^{4}}\\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\frac{\left| ( 4\mathcal{O}( A_{k}) -1) c\Delta t-( \Delta x)^{4}\right| }{4( \Delta x)^{2}} +\frac{\left[( \Delta x)^{4} +c\Delta t\right]^{2}}{16( \Delta x)^{4}}\\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\frac{1}{4( \Delta x)^{2}}\left\{\left| ( 4\mathcal{O}( A_{k}) -1) c\Delta t-( \Delta x)^{4}\right| +\frac{\left[( \Delta x)^{4} +c\Delta t\right]^{2}}{4( \Delta x)^{2}}\right\}
    \end{array}
\end{align}

aquí el segundo término dentro de las llaves es el que nos da el orden de magnitud más grande para el error, entonces,

\begin{align}
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) \approx \frac{\left[( \Delta x)^{8} +2c( \Delta x)^{4} \Delta t+( \Delta t)^{2}\right]}{16( \Delta x)^{4}} \\

    \mathrm{si} \ \Delta x,\Delta t\ll 1\Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) \approx \left[( \Delta x)^{4} +\Delta t+\left(\frac{\Delta t}{( \Delta x)^{2}}\right)^{2}\right] \Rightarrow \boxed{\mathcal{O}( \epsilon _{\mathrm{imp}})\approx \left(\frac{\Delta t}{( \Delta x)^{2}}\right)^{2} \to \mathrm{a \space 1er \space orden \space}} 
\end{align}

donde hemos despreciado los primeros dos términos del corchete por sobre el tercero.

Si, en cambio, retenemos términos hasta segundo orden en $( \lambda _{k} \Delta t)$ para los desarrollos de Taylor tendremos,

\begin{align}
    \begin{array}{l}
    \exp\left( -\frac{i\lambda _{k} t}{\hbar }\right) =\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) +\frac{1}{2}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right)^{2} +\mathcal{O}\left(\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right)^{3}\right)\\
    \\
    \frac{1}{( s_{1} -is_{2})} =\frac{is_{2}}{( s_{1})^{2}} -\frac{( s_{2})^{2}}{( s_{1})^{3}} +\mathcal{O}\left(\left(\frac{is_{2}}{( s_{1})^{2}}\right)^{3}\right) ,\ ( s_{i} \in \mathbb{R})
    \end{array}
\end{align}

entonces,

\begin{align}
    \begin{array}{l}
    \epsilon _{\mathrm{imp}} =\left| A_{k}\left[\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right) +\frac{1}{2}\left( -\frac{i\lambda _{k} \Delta t}{\hbar }\right)^{2}\right] -\dotsc \right| \\
    \left| \dotsc -\frac{\left\{4+i\left[\tfrac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\tfrac{\hbar \Delta t\lambda _{k}}{m}\right]\right\}\left\{i\left[\tfrac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\tfrac{\hbar \Delta t\lambda _{k}}{m}\right] -\frac{1}{4}\left[\tfrac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\tfrac{\hbar \Delta t\lambda _{k}}{m}\right]^{2}\right\}}{( 4)^{2}}\right| 
    \end{array}
\end{align}

\begin{align}
    \begin{array}{l}
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\left[\left( i\frac{c\Delta t}{( \Delta x)^{2}}\right) -\frac{1}{2}\left(\frac{c\Delta t}{( \Delta x)^{2}}\right)^{2}\right] -\frac{\left\{4+i\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]\right\}\left\{i\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right] -\frac{1}{4}\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]^{2}\right\}}{16}\right| \\
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) =\left| \mathcal{O}( A_{k})\left[\left( i\frac{c\Delta t}{( \Delta x)^{2}}\right) -\frac{1}{2}\left(\frac{c\Delta t}{( \Delta x)^{2}}\right)^{2}\right] -\frac{\left\{-2\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]^{2} +\frac{i}{4}\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]\left\{16-\left[( \Delta x)^{2} +\frac{c\Delta t}{( \Delta x)^{2}}\right]^{2}\right\}\right\}}{16}\right| 
    \end{array}
\end{align}

y los términos de mayor orden serán

\begin{align}
    \Rightarrow \mathcal{O}( \epsilon _{\mathrm{imp}}) \approx \left[( \Delta x)^{8} +\left(\frac{\Delta t}{( \Delta x)^{2}}\right)^{3}\right] \Rightarrow \boxed{\mathcal{O}( \epsilon _{\mathrm{imp}}) \approx \left(\frac{\Delta t}{( \Delta x)^{2}}\right)^{3} \to \mathrm{a \space 2do \space orden \space}}
\end{align}

Entonces, como el método implícito es un método de 2do orden, tendremos que considerar el error reteniendo términos del desarrollo de Taylor hasta 2do orden, lo cual nos dice que el error escalea como el último resultado. Este último resultado fue el que hemos visto en el curso, sin embargo, el análisis anterior permite notar los ordenes que debemos retener en los desarrollos para reproducir dicho resultado.

Finalmente, notemos que el resultado anterior es válido para las regiones de convergencia de los desarrollos en serie de Taylor, es decir,

\begin{align}
    \begin{array}{l}
    \exp( x)\rightarrow x\ll 1\Rightarrow \left| \frac{-i\lambda _{k} \Delta t}{\hbar }\right| =\left| \frac{ic\Delta t}{\hbar ( \Delta x)^{2}}\right| \ll 1\Rightarrow \begin{cases}
    \hbar =1\\
    c=1
    \end{cases} \Rightarrow \boxed{\Delta t\ll ( \Delta x)^{2}}\\
    \frac{1}{( 1-z)}\rightarrow | z| < 1\Rightarrow \left| \frac{is_{2}}{s_{1}}\right| =\left| \frac{i\left[\frac{m\omega ^{2}( x-x_{a})^{2}}{\hbar } -\frac{\hbar \Delta t\lambda _{k}}{m}\right]}{4}\right| =\left| \frac{i\left[\frac{m\omega ^{2}( \Delta x)^{2}}{\hbar } +\frac{\hbar c\Delta t}{m( \Delta x)^{2}}\right]}{4}\right| < 1\\
    (\mathrm{a.u.}) \Rightarrow \begin{cases}
    \hbar =1\\
    c=1\\
    m=1\\
    \omega =1
    \end{cases} \Rightarrow \frac{( \Delta x)^{4} +\Delta t}{( \Delta x)^{2}} < 4\Rightarrow ( \Delta x)^{4} -4( \Delta x)^{2} +\Delta t< 0
    \end{array}
\end{align}

\begin{align}
    ( \Delta x)^{2} =\Lambda \Rightarrow \Lambda ^{2} -4\Lambda +\Delta t=0\Leftrightarrow \Lambda _{\pm } =( \Delta x)^{2} =\begin{cases}
    2+\sqrt{4-\Delta t}\\
    2-\sqrt{4-\Delta t}
    \end{cases} \Rightarrow \Delta x_{\pm \pm } =\begin{cases}
    2+\sqrt{4-\Delta t}\\
    2-\sqrt{4-\Delta t}\\
    -2-\sqrt{4-\Delta t}\\
    -2+\sqrt{4-\Delta t}
    \end{cases}
\end{align}

y tenemos cuatro valores posibles para $\Delta x$. Si proponemos que la discretización temporal utilizada en el método será tal que $0< \Delta t< 4\Rightarrow 0< \sqrt{4-\Delta t} < 2$, sin embargo, para que nos encontremos dentro de las regiones de convergencia deberemos cumplir que,

\begin{align}
    \begin{array}{l}
    \Rightarrow \left[( \Delta x)^{4} -4( \Delta x)^{2} +\Delta t\right] < 0\\
    \Rightarrow ( \Delta x-\Delta x_{++})( \Delta x-\Delta x_{+-})( \Delta x-\Delta x_{--})( \Delta x-\Delta x_{-+}) < 0
    \end{array}
\end{align}

\begin{align}
    \Rightarrow \left[ \Delta x-\underbrace{\left( 2+\sqrt{4-\Delta t}\right)}_{ >0}\right]\left[ \Delta x-\underbrace{\left( 2-\sqrt{4-\Delta t}\right)}_{ >0}\right]\left[ \Delta x-\underbrace{\left( -2-\sqrt{4-\Delta t}\right)}_{< 0}\right]\left[ \Delta x-\underbrace{\left( -2+\sqrt{4-\Delta t}\right)}_{< 0}\right] < 0
\end{align}

y como la discretización espacial siempre será tal que $\Delta x >0$ entonces deberá cumplirse que,

\begin{align}
    \boxed{[ 0< \Delta t< 4] \land \left[\left( 2-\sqrt{4-\Delta t}\right) < \Delta x< \left( 2+\sqrt{4-\Delta t}\right)\right]}
\end{align}

entonces, si por ejemplo, usamos que $\Delta x=0.1$, para cumplir con la primera condición tendremos que proponer que $\frac{\Delta t}{( \Delta x)^{2}} \ll 1$ y como regla general usaremos que $\Delta t=\alpha ( \Delta x)^{2} ;\ 0< \alpha \ll 1$ entonces, por la última condición que debemos cumplir tendremos,

\begin{align}
    \left( 2-\sqrt{4-\alpha ( \Delta x)^{2}}\right) < \Delta x< \left( 2+\sqrt{4-\alpha ( \Delta x)^{2}}\right) ;\ \alpha \ll 1
\end{align}

y si graficamos estas regiones tendremos,

entonces vemos que la condición $\Delta x >\left( 2-\sqrt{4-\alpha ( \Delta x)^{2}}\right)$ se cumple $\forall \alpha  >0$ y, a medida que disminuimos $\alpha$ obtenemos un rango de validez más amplio para $\Delta x$ posibles. Entonces, proponiendo la condición $\Delta t=\alpha ( \Delta x)^{2}$ con $0< \alpha \ll 1$ cumpliremos también la segunda condición, matemáticamente esto sería,

\begin{align}
    \begin{array}{l}
    \Delta x< \left( 2+\sqrt{4-\alpha ( \Delta x)^{2}}\right) \Rightarrow ( \Delta x-2)^{2} < \left| 4-\alpha ( \Delta x)^{2}\right| =4-\alpha ( \Delta x)^{2}\\
    \Rightarrow ( \Delta x)^{2} +4\Delta x+4-4+\alpha ( \Delta x)^{2} < 0\Rightarrow ( 1+\alpha )( \Delta x)^{2} +4\Delta x >0\\
    \Rightarrow \Delta x >-\frac{4}{( 1+\alpha )} \ \Rightarrow (\mathrm{se\ cumple} \ \forall \ \alpha  >0)\\
    \\
    \Delta x >\left( 2-\sqrt{4-\alpha ( \Delta x)^{2}}\right) \Rightarrow ( 2-\Delta x)^{2} < \left| 4-\alpha ( \Delta x)^{2}\right| =4-\alpha ( \Delta x)^{2}\\
    \Rightarrow ( \Delta x)^{2} -4\Delta x+4-4+\alpha ( \Delta x)^{2} < 0\Rightarrow ( 1+\alpha )( \Delta x)^{2} -4\Delta x< 0\\
    \Rightarrow \Delta x< \frac{4}{( 1+\alpha )}
    \end{array}
\end{align}

## 3. Segundo método de resolución

Aquí la idea es resolver el problema encontrando las autoenergías y autoestados del hamiltoniano completo del sistema $\hat{H}( x)$, luego recordando la definición de operador evolución uitaria de la mecánica cuántica podremos evolucionar trivialmente el sistema.

### 3.1. Problema de autovalores del hamiltoniano $\hat{h}_{x_{0}}$ unidimensional (resolución de partes real e imaginaria de autoestados por separado)

En primera intancia buscamos diagonalizar el operador $\hat{h}_{x_{a}}( x)$ entonces queremos resolver la ec. de Schrödinger independiente del tiempo de la forma (ec. de autovalores),

\begin{align}
    \begin{array}{l}
    \hat{h}_{x_{a}}( x) \phi ( x) =\lambda \phi ( x) \Rightarrow \left[ -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2}}{\partial x^{2}} +\frac{1}{2} m\omega ^{2}( x-x_{a})^{2}\right] \phi ( x) =\lambda \phi ( x)\\
    \Rightarrow \left[ -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2}}{\partial x^{2}} +\frac{1}{2} m\omega ^{2}( x-x_{a})^{2}\right]\left(\overbrace{\mathbb{R}\mathrm{e}\{\phi ( x)\}}^{=u_{1}( x)} +i\overbrace{\Im \mathrm{m}\{\phi ( x)\}}^{=u_{2}( x)}\right) =\lambda (\mathbb{R}\mathrm{e}\{\phi ( x)\} +i\Im \mathrm{m}\{\phi ( x)\})
    \end{array}
\end{align}

\begin{align}
    \begin{array}{l}
    \Rightarrow \left[ -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2}}{\partial x^{2}} +\frac{1}{2} m\omega ^{2}( x-x_{a})^{2}\right]\hat{I}\begin{pmatrix}
    u_{1}( x)\\
    i\cdotp u_{2}( x)
    \end{pmatrix} =\lambda \hat{I}\begin{pmatrix}
    u_{1}( x)\\
    i\cdotp u_{2}( x)
    \end{pmatrix}\\
    \Rightarrow \left[ -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2}}{\partial x^{2}} +\frac{1}{2} m\omega ^{2}( x-x_{a})^{2}\right]\hat{I}\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)
    \end{pmatrix} =\lambda \hat{I}\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)
    \end{pmatrix}
    \end{array}
\end{align}

donde hemos multiplicado la ecuación para la parte imaginaria m.a.m. por $(-i)$ entonces, el sistema se reduce a una ecuación de Sturm-Liouville con parámetros $p( x) ,q( x) ,r( x)$ defnidos como,

\begin{align}
    \begin{cases}
    p( x) =\frac{\hbar ^{2}}{2m}\\
    q( x) =\frac{1}{2} m\omega ^{2}( x-x_{a})^{2}\\
    r( x) =1
    \end{cases} \Rightarrow -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q( x)\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)
    \end{pmatrix} =\lambda r( x)\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)
    \end{pmatrix}
\end{align}

Las formas bilineales para considerar el problema débil de autovalores serán de la forma,

\begin{align}
    \begin{array}{l}
    \begin{cases}
    a_{1}( u_{1} ,v_{1}) := \int _{\Omega }[ p( x)( \nabla v_{1} \cdotp \nabla u_{1}) +q( x)( v_{1} u_{1})] d\Omega \\
    a_{2}( u_{2} ,v_{2}) := \int _{\Omega }[ p( x)( \nabla v_{2} \cdotp \nabla u_{2}) +q( x)( v_{2} u_{2})] d\Omega 
    \end{cases}\\
    a(( u_{1} ,u_{2}) ,( v_{1} ,v_{2})) := a_{1}( u_{1} ,v_{1}) +a_{2}( u_{2} ,v_{2})\\
    \\
    \begin{cases}
    b_{1}( u_{1} ,v_{1}) := \int _{\Omega } r( x)( v_{1} u_{1}) d\Omega ; & b_{2}( u_{2} ,v_{2}) := \int _{\Omega } r( x)( v_{2} u_{2}) d\Omega 
    \end{cases}\\
    b(( u_{1} ,u_{2}) ,( v_{1} ,v_{2})) := b_{1}( u_{1} ,v_{1}) +b_{2}( u_{2} ,v_{2})
    \end{array}
\end{align}

### 3.2. Problema de autovalores del hamiltoniano del sistema completo $\hat{H}$ de dimensión 2 (resolución de autoestados en un espacio complejo)

Buscamos diagonalizar el operador $\hat{H}(x)$ entonces queremos resolver la ec. de Schrödinger independiente del tiempo de la forma (ec. de autovalores):

\begin{align}
    \begin{array}{l}
    \hat{H}( x)\vec{\phi }( x) =\lambda \vec{\phi }( x) \Rightarrow \begin{pmatrix}
    h_{x_{a}}( x) & \gamma \\
    \gamma  & h_{x_{b}}( x)
    \end{pmatrix}\vec{\phi }( x) =\lambda \vec{\phi }( x)\\
    \Rightarrow \begin{cases}
    -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2} \phi _{1}( x)}{\partial x^{2}} +\left[\frac{1}{2} m\omega ^{2}( x-x_{a})^{2} \phi _{1}( x) +\gamma \phi _{2}( x)\right] =\lambda \phi _{1}( x) & \\
    -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2} \phi _{2}( x)}{\partial x^{2}} +\left[\frac{1}{2} m\omega ^{2}( x-x_{b})^{2} \phi _{2}( x) +\gamma \phi _{1}( x)\right] =\lambda \phi _{2}( x) & 
    \end{cases}
    \end{array}
\end{align}

entonces según nuestro problema particular tenemos un sistema de ecuaciones de tipo Sturm-Liovulle con parámetros $p( x) ,q( x) ,r( x)$ definidos como,

\begin{align}
    \begin{array}{l}
    \begin{cases}
    p( x) =\frac{\hbar ^{2}}{2m}\\
    q_{1}( x) =\frac{1}{2} m\omega ^{2}( x-x_{a})^{2} ;q_{2}( x) =\frac{1}{2} m\omega ^{2}( x-x_{b})^{2}\\
    r( x) =1;\ s( x) =\gamma 
    \end{cases}\\
    \\
    \Rightarrow \begin{pmatrix}
    \left[ -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q_{1}( x)\right] & s( x)\\
    s( x) & \left[ -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q_{2}( x)\right]
    \end{pmatrix}\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)
    \end{pmatrix} =\lambda r( x)\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)
    \end{pmatrix}\\
    \Rightarrow \begin{cases}
    \left[ -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q_{1}( x)\right] u_{1}( x) +s( x) u_{2}( x) =\lambda r( x) u_{1}( x)\\
    \left[ -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q_{2}( x)\right] u_{2}( x) +s( x) u_{1}( x) =\lambda r( x) u_{2}( x)
    \end{cases}
    \end{array}
\end{align}

Las formas bilineales dobles para considerar el problema débil de autovalores serán de la forma

\begin{align}
    \begin{array}{l}
    \begin{cases}
    a_{1}(( u_{1} ,u_{2}) ,v_{1}) := \int _{\Omega }[ p( x)( \nabla v_{1} \cdotp \nabla u_{1}) +q_{1}( x) v_{1} u_{1} +s( x) v_{1} u_{2}] d\Omega \\
    a_{2}(( u_{2} ,u_{1}) ,v_{2}) := \int _{\Omega }[ p( x)( \nabla v_{2} \cdotp \nabla u_{2}) +q_{2}( x) v_{2} u_{2} +s( x) v_{2} u_{1}] d\Omega 
    \end{cases}\\
    \\
    \Rightarrow a(( u_{1} ,u_{2}) ,( v_{1} ,v_{2})) := a_{1}(( u_{1} ,u_{2}) ,v_{1}) +a_{2}(( u_{2} ,u_{1}) ,v_{2})\\
    \\
    \begin{cases}
    b_{1}( u_{1} ,v_{1}) := \int _{\Omega } r( x)( u_{1} v_{1}) d\Omega ; & b_{2}( u_{2} ,v_{2}) := \int _{\Omega } r( x)( u_{2} v_{2}) d\Omega 
    \end{cases}\\
    \Rightarrow b(( u_{1} ,u_{2}) ,( v_{1} ,v_{2})) := b_{1}( u_{1} ,v_{1}) +b_{2}( u_{2} ,v_{2})
    \end{array}
\end{align}

### 3.3. Problema de autovalores del hamiltoniano del sistema completo $\hat{H}$ de dimensión 2 (resolución de partes real e imaginaria de autoestados por separado)

Buscamos diagonalizar el operador $\hat{H}( x)$ entonces queremos resolver la ec. de Schrödinger independiente del tiempo de la forma (ec. de autovalores):

\begin{align}
    \begin{array}{l}

\end{align}

\begin{align}
    \Rightarrow \begin{cases}

donde $u_{1} =\mathbb{R}\mathrm{e}\{\phi _{1}( x)\} ;u_{2} =\Im \mathrm{m}\{\phi _{1}( x)\} ;u_{3} =\mathbb{R}\mathrm{e}\{\phi _{2}( x)\} ;u_{4} =\Im \mathrm{m}\{\phi _{2}( x)\}$

\begin{align}
    \begin{array}{l}
    \Rightarrow \left[ -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2}}{\partial x^{2}}\hat{I} +\frac{1}{2} m\omega ^{2}\begin{pmatrix}
    ( x-x_{a})^{2} & 0 & 0 & 0\\
    0 & ( x-x_{a})^{2} & 0 & 0\\
    0 & 0 & ( x-x_{b})^{2} & 0\\
    0 & 0 & 0 & ( x-x_{b})^{2}
    \end{pmatrix}\right]\begin{pmatrix}
    u_{1}( x)\\
    i\cdotp u_{2}( x)\\
    u_{3}( x)\\
    i\cdotp u_{4}( x)
    \end{pmatrix} +\gamma \hat{I}\begin{pmatrix}
    u_{3}( x)\\
    i\cdotp u_{4}( x)\\
    u_{1}( x)\\
    i\cdotp u_{2}( x)
    \end{pmatrix} =\lambda \hat{I}\begin{pmatrix}
    u_{1}( x)\\
    i\cdotp u_{2}( x)\\
    u_{3}( x)\\
    i\cdotp u_{4}( x)
    \end{pmatrix}\\
    \Rightarrow \left[ -\frac{\hbar ^{2}}{2m}\frac{\partial ^{2}}{\partial x^{2}} +\frac{1}{2} m\omega ^{2}( x-x_{a})^{2}\right]\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)\\
    u_{3}( x)\\
    u_{4}( x)
    \end{pmatrix} +\gamma \begin{pmatrix}
    u_{3}( x)\\
    u_{4}( x)\\
    u_{1}( x)\\
    u_{2}( x)
    \end{pmatrix} =\lambda \begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)\\
    u_{3}( x)\\
    u_{4}( x)
    \end{pmatrix}
    \end{array}
\end{align}

donde hemos multiplicado las ecuaciones de las partes imaginarias m.a.m. por $(-i)$ entonces, el sistema se reduce a una ecuación de Sturm-Liouville con parámetros $p( x) ,q_{i}( x) ,s( x) ,r( x)$ definidos como,

\begin{align}
    \begin{cases}
    p( x) =\frac{\hbar ^{2}}{2m}\\
    q_{1}( x) =\frac{1}{2} m\omega ^{2}( x-x_{a})^{2} ;q_{2}( x) =\frac{1}{2} m\omega ^{2}( x-x_{b})^{2}\\
    r( x) =1;\ s( x) =\gamma 
    \end{cases}
\end{align}

\begin{align}
    \Rightarrow \left[ -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right)\hat{I} +\begin{pmatrix}
    q_{1}( x) & 0 & 0 & 0\\
    0 & q_{1}( x) & 0 & 0\\
    0 & 0 & q_{2}( x) & 0\\
    0 & 0 & 0 & q_{2}( x)
    \end{pmatrix}\right]\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)\\
    u_{3}( x)\\
    u_{4}( x)
    \end{pmatrix} +s( x)\hat{I}\begin{pmatrix}
    u_{3}( x)\\
    i\cdotp u_{4}( x)\\
    u_{1}( x)\\
    i\cdotp u_{2}( x)
    \end{pmatrix} =\lambda r( x)\hat{I}\begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)\\
    u_{3}( x)\\
    u_{4}( x)
    \end{pmatrix}
\end{align}

y esto es equivalente al siguiente sistema (en formato matricial) $A\vec{u} =\lambda \vec{u}$ donde

\begin{align}
    A := \begin{pmatrix}
    -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q_{1}( x) & 0 & s( x) & 0\\
    0 & -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q_{1}( x) & 0 & s( x)\\
    s( x) & 0 & -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q_{2}( x) & 0\\
    0 & s( x) & 0 & -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q_{2}( x)
    \end{pmatrix}
\end{align}

\begin{align}
    \vec{u} := \begin{pmatrix}
    u_{1}( x)\\
    u_{2}( x)\\
    u_{3}( x)\\
    u_{4}( x)
    \end{pmatrix}
\end{align}

Las formas bilineales para considerar el problema débil de autovalores serán de la forma,

\begin{align}
    \begin{array}{l}
    \begin{cases}
    a_{1}(( u_{1} ,u_{3}) ,v_{1}) := \int _{\Omega }[ p( \nabla v_{1} \cdotp \nabla u_{1}) +q_{1} v_{1} u_{1} +sv_{1} u_{3}] d\Omega \\
    a_{2}(( u_{2} ,u_{4}) ,v_{2}) := \int _{\Omega }[ p( \nabla v_{2} \cdotp \nabla u_{2}) +q_{1} v_{2} u_{2} +sv_{2} u_{4}] d\Omega \\
    a_{3}(( u_{3} ,u_{1}) ,v_{3}) := \int _{\Omega }[ p( \nabla v_{3} \cdotp \nabla u_{3}) +q_{2} v_{3} u_{3} +sv_{3} u_{1}] d\Omega \\
    a_{4}(( u_{4} ,u_{2}) ,v_{4}) := \int _{\Omega }[ p( \nabla v_{4} \cdotp \nabla u_{4}) +q_{2} v_{4} u_{4} +sv_{4} u_{2}] d\Omega 
    \end{cases}\\
    a(( u_{1} ,u_{2} ,u_{3} ,u_{4}) ,( v_{1} ,v_{2} ,v_{3} ,v_{4})) := \sum\nolimits _{i=1}^{4} a_{i}(( u_{i} ,u_{j}) ,v_{i})\\
    \\
    \begin{cases}
    b_{1}(( u_{1} ,v_{1})) := \int _{\Omega } rv_{1} u_{1} d\Omega ; & b_{2}(( u_{2} ,v_{2})) := \int _{\Omega } rv_{2} u_{2} d\Omega \\
    b_{3}(( u_{3} ,v_{3})) := \int _{\Omega } rv_{3} u_{3} d\Omega ; & \ b_{4}(( u_{4} ,v_{4})) := \int _{\Omega } rv_{4} u_{4} d\Omega 
    \end{cases}\\
    b(( u_{1} ,u_{2} ,u_{3} ,u_{4}) ,( v_{1} ,v_{2} ,v_{3} ,v_{4})) := \sum\nolimits _{i=1}^{4} b_{i}(( u_{i} ,v_{i}))
    \end{array}
\end{align}

entonces todo se resume a resolver un problema del tipo,

\begin{align}
    a(( u_{1} ,u_{2} ,u_{3} ,u_{4}) ,( v_{1} ,v_{2} ,v_{3} ,v_{4})) =\lambda b(( u_{1} ,u_{2} ,u_{3} ,u_{4}) ,( v_{1} ,v_{2} ,v_{3} ,v_{4}))
\end{align}

## Anexo I: Análisis dimensional

### I. Adimensionalizando el problema de autovalores

Partimos de un problema de Sturm-Liouville de la forma

\begin{align}
    \left[ -\frac{d}{dx}\left( p( x)\frac{d}{dx}\right) +q( x)\right] u( x) =\lambda r( x) u( x)
\end{align}

definiendo los siguientes parámetros de espacio, tiempo y masa adimensionales,

\begin{align}
    x_{ad} := \frac{x}{c_{x}} ;\ t_{ad} := \frac{t}{c_{t}} ;\ m_{ad} := \frac{m}{c_{m}}
\end{align}

y, usando regla de la cadena, podemos obtener el siguiente resultado para los operadores diferenciales espaciales y temporales:

\begin{align}
    \begin{array}{l}
    \frac{\partial }{\partial x} =\frac{\partial }{\partial x_{ad}}\frac{\partial x_{ad}}{\partial x} =\frac{1}{c_{x}}\frac{\partial }{\partial x_{ad}}\\
    \frac{\partial ^{2}}{\partial x^{2}} =\left[\frac{\partial ^{2}}{\partial ( x_{ad})^{2}}\left(\frac{\partial x_{ad}}{\partial x}\right)^{2} +\frac{\partial }{\partial x_{ad}}\cancel{\frac{\partial ^{2} x_{ad}}{\partial x^{2}}}\right] =\frac{1}{( c_{x})^{2}}\frac{\partial ^{2}}{\partial ( x_{ad})^{2}}\\
    \frac{\partial }{\partial t} =\frac{\partial }{\partial t_{ad}}\frac{\partial t_{ad}}{\partial t} =\frac{1}{c_{t}}\frac{\partial }{\partial t_{ad}}
    \end{array}
\end{align}

Por otro lado, debemos adimensionalizar otras cantidades que forman parte del problema de autovalores como ser;

\begin{align}
    \begin{array}{l}
    [ \hbar _{ad}] := \frac{[ \hbar ]}{[ c_{\hbar }]} \Rightarrow \frac{\left[ ML^{2} T^{-1}\right]}{[ c_{\hbar }]} =[ 1] \Rightarrow \hbar _{ad} =\frac{\hbar }{c_{m}( c_{x})^{2}( c_{t})^{-1}}\rightarrow \left(\begin{array}{ c }
    \mathrm{constante\ de}\\
    \mathrm{Planck}
    \end{array}\right)\\
    [ \omega _{ad}] := \frac{[ \omega ]}{[ c_{\omega }]} =\frac{\left[ T^{-1}\right]}{[ c_{\omega }]} =[ 1] \Rightarrow \omega _{ad} =\frac{\omega }{( c_{t})^{-1}}\rightarrow \left(\begin{array}{ c }
    \mathrm{frecuencia}\\
    \mathrm{angular}
    \end{array}\right)\\
    [ \gamma _{ad}] := \frac{[ \gamma ]}{[ c_{\gamma }]} =\frac{\left[ ML^{2} T^{-2}\right]}{[ c_{\gamma }]} =[ 1] \Rightarrow \gamma _{ad} =\frac{\gamma }{c_{m}( c_{x})^{2}( c_{t})^{-2}}\rightarrow \left(\begin{array}{ c }
    \mathrm{factor\ de}\\
    \mathrm{acoplamiento}
    \end{array}\right)\\
    [ \lambda _{ad}] := \left[\frac{\lambda }{c_{\lambda }}\right] =\frac{\left[ ML^{2} T^{-2}\right]}{[ c_{\lambda }]} =[ 1] \Rightarrow \lambda _{ad} =\frac{\lambda }{c_{m}( c_{x})^{2}( c_{t})^{-2}}\rightarrow \left(\begin{array}{ c }
    \mathrm{auto}\\
    \mathrm{energía}
    \end{array}\right)
    \end{array}
\end{align}

donde hemos hecho uso de que $[\mathrm{energía}] =\left[ ML^{2} T^{-2}\right]$. Si hacemos uso de las definiciones de las funciones $p( x) ,q( x) ,r( x)$ para nuestro problema particular entonces, podremos obtener expresiones adimensionales de ellas de la siguiente manera,

\begin{align}
    \begin{array}{l}
    p( x) := \frac{\hbar ^{2}}{2m} =\frac{\left[ \hbar _{ad} c_{m}( c_{x})^{2}( c_{t})^{-1}\right]^{2}}{2( m_{ad} c_{m})} =p_{ad}( x_{ad}) c_{p}\\
    \therefore p_{ad}( x_{ad}) =\frac{( \hbar _{ad})^{2}}{2m_{ad}} ;c_{p} =c_{m}( c_{x})^{4}( c_{t})^{-2}\\
    \\
    q( x) := \left[\frac{m\omega ^{2}( x-x_{a})^{2}}{2} +\gamma \right]\\
    \Rightarrow q( x) =\frac{( m_{ad} c_{m})\left( \omega _{ad}( c_{t})^{-1}\right)^{2}[ c_{x}( x_{ad} -x_{a,ad})]^{2}}{2} +\gamma _{ad}\left[ c_{m}( c_{x})^{2}( c_{t})^{-2}\right] =q_{ad}( x_{ad}) c_{q}\\
    \therefore q_{ad}( x_{ad}) =\left[\frac{m_{ad}( \omega _{ad})^{2}( x_{ad} -x_{a,ad})^{2}}{2} +\gamma _{ad}\right] ;\ c_{q} =c_{m}( c_{x})^{2}( c_{t})^{-2}\\
    \\
    r( x) := 1=c_{r} r_{ad}( x_{ad}) \therefore r_{ad}( x_{ad}) =1;c_{r} =1
    \end{array}
\end{align}

luego, reemplazando todos estos resultados en el problema de autovalores original, podremos llegar a una expresión adimensional de la forma,

\begin{align}
    \begin{array}{l}
    \left[ -\frac{1}{c_{x}}\frac{\partial }{\partial x_{ad}}\left( c_{p} p_{ad}( x_{ad})\frac{1}{c_{x}}\frac{\partial }{\partial x_{ad}}\right) +c_{q} q_{ad}( x_{ad})\right] u_{ad}( x_{ad}) =\dotsc \\
    =\dotsc \left[ c_{m}( c_{x})^{2}( c_{t})^{-2} \lambda _{ad}\right][ c_{r} r_{ad}( x_{ad})] u_{ad}( x_{ad})\\
    \Rightarrow \left[ -\frac{\partial }{\partial x_{ad}}\left(\frac{c_{p}}{( c_{x})^{2}} p_{ad}( x_{ad})\frac{\partial }{\partial x_{ad}}\right) +c_{q} q_{ad}( x_{ad})\right] u_{ad}( x_{ad}) =\dotsc \\
    =\dotsc \left[ c_{m}( c_{x})^{2}( c_{t})^{-2} \lambda _{ad}\right][ c_{r} r_{ad}( x_{ad})] u_{ad}( x_{ad})\\
    \Rightarrow c_{m}( c_{x})^{2}( c_{t})^{-2}\left[ -\frac{\partial }{\partial x_{ad}}\left( p_{ad}( x_{ad})\frac{\partial }{\partial x_{ad}}\right) +q_{ad}( x_{ad})\right] u_{ad}( x_{ad}) =\dotsc \\
    =\dotsc \left[ c_{m}( c_{x})^{2}( c_{t})^{-2} \lambda _{ad}\right][ c_{r} r_{ad}( x_{ad})] u_{ad}( x_{ad})\\
    \Rightarrow \boxed{\left[ -\frac{\partial }{\partial x_{ad}}\left( p_{ad}( x_{ad})\frac{\partial }{\partial x_{ad}}\right) +q_{ad}( x_{ad})\right] u_{ad}( x_{ad}) =\lambda _{ad} r_{ad}( x_{ad}) u_{ad}( x_{ad})}
    \end{array}
\end{align}

donde hemos usado que $[ c_{r}] =[ 1]$. Entonces, como conclusión podemos decir que, independientemente de las unidades de $x,t,m$ podremos resolver el problema de autovalores anterior y, en caso de requerirlo, re-escalear los resultados finales para obtener resultados dimensionales.

### II. Adimensionalizando la ecuación de Schrödinger (evolución temporal)

Partimos de la siguiente ecuación para la evolución temporal que tenemos para nuestro problema particular,

\begin{align}
    \begin{array}{l}
    \frac{\partial \psi _{a}( x,t)}{\partial t} =\alpha \frac{\partial ^{2} \psi _{a}( x,t)}{\partial x^{2}} +\alpha _{i}( x) \psi _{a}( x,t) +\beta \psi _{b}( x,t)\\
    \mathrm{donde} \ \alpha := \frac{i\hbar }{2m} ;\alpha _{i}( x) =\frac{-im\omega ^{2}( x-x_{i})^{2}}{2\hbar } ;\beta =\frac{-i\gamma }{\hbar }
    \end{array}
\end{align}

y utilizando los parámetros $x_{ad} ,t_{ad}$ definidos anteriormente tendremos,

\begin{align}
    \frac{1}{c_{t}}\frac{\partial \psi _{a,ad}( x_{ad} ,t_{ad})}{\partial t_{ad}} =\frac{\alpha }{( c_{x})^{2}}\frac{\partial ^{2} \psi _{a,ad}( x_{ad} ,t_{ad})}{\partial ( x_{ad})^{2}} +\alpha _{1}( x) \psi _{a,ad}( x_{ad} ,t_{ad}) +\beta \psi _{b,ad}( x_{ad} ,t_{ad})
\end{align}

y usando las definiciones particulares de las funciones $\alpha ,\alpha _{i}( x) ,\beta $ tendremos expresiones adimensionales para ellas de la forma;

\begin{align}
    \begin{array}{l}
    \alpha =\frac{i\hbar }{2m} =\frac{i\left[ \hbar _{ad} c_{m}( c_{x})^{2}( c_{t})^{-1}\right]}{2( m_{ad} c_{m})} =c_{\alpha } \alpha _{ad} \therefore \alpha _{ad} =\frac{i\hbar _{ad}}{2m_{ad}} ;c_{\alpha } =( c_{x})^{2}( c_{t})^{-1}\\
    \\
    \alpha _{i}( x) =\frac{-im\omega ^{2}( x-x_{i})^{2}}{2\hbar } =\frac{-i( m_{ad} c_{m})\left[ \omega _{ad}( c_{t})^{-1}\right]^{2}[ c_{x}( x_{ad} -x_{i,ad})]^{2}}{2\left[ \hbar _{ad} c_{m}( c_{x})^{2}( c_{t})^{-1}\right]} =c_{\alpha _{i}} \alpha _{i,ad}( x_{ad})\\
    \therefore \alpha _{i,ad}( x_{ad}) =\frac{-im_{ad}( \omega _{ad})^{2}( x_{ad} -x_{i,ad})^{2}}{2\hbar _{ad}} ;\ c_{\alpha _{i}} =( c_{t})^{-1}\\
    \\
    \beta =\frac{-i\gamma }{\hbar } =\frac{-i\left[ \gamma _{ad} c_{m}( c_{x})^{2}( c_{t})^{-2}\right]}{\left[ \hbar _{ad} c_{m}( c_{x})^{2}( c_{t})^{-1}\right]} =c_{\beta } \beta _{ad} \therefore \beta _{ad} =\frac{-i\gamma _{ad}}{\hbar _{ad}} ;\ c_{\beta } =( c_{t})^{-1}
    \end{array}
\end{align}

entonces reemplazando todos los resultados en la ecuación de evolución temporal tendremos,

\begin{align}
    \begin{array}{l}
    ( c_{t})^{-1}\frac{\partial \psi _{a,ad}( x_{ad} ,t_{ad})}{\partial t_{ad}} =\frac{( c_{x})^{2}( c_{t})^{-1} \alpha _{ad}}{( c_{x})^{2}}\frac{\partial ^{2} \psi _{a,ad}( x_{ad} ,t_{ad})}{\partial ( x_{ad})^{2}} +\dotsc \\
    \dotsc +( c_{t})^{-1} \alpha _{1,ad}( x) \psi _{a,ad}( x_{ad} ,t_{ad}) +( c_{t})^{-1} \beta _{ad} \psi _{b,ad}( x_{ad} ,t_{ad})\\
    \Rightarrow \boxed{\frac{\partial \psi _{a,ad}( x_{ad} ,t_{ad})}{\partial t_{ad}} =\alpha _{ad}\frac{\partial ^{2} \psi _{a,ad}( x_{ad} ,t_{ad})}{\partial ( x_{ad})^{2}} +\alpha _{1,ad}( x) \psi _{a,ad}( x_{ad} ,t_{ad}) +\beta _{ad} \psi _{b,ad}( x_{ad} ,t_{ad})}\\
    \end{array}
\end{align}

Y, nuevamente, como conlcusión podemos decir que, independientemente de las unidades de $x,t,m$ podremos resolver el problema de evolución temporal anterior y re-escalear los resultados finales para obtener resultados dimensionales.

### III. Elección del sistema de unidades

Numéricamente, utilizando el [sistema atómico de unidades](https://en.wikipedia.org/wiki/Hartree_atomic_units) (Hartree atomic units) entonces,

\begin{align}
    \begin{array}{l}
    c_{\hbar } =\hbar =1.054571\cdotp 10^{-34}[ J\cdotp s]\rightarrow (\mathrm{Planck\ constant})\\
    c_{x} =a_{0} =5.291772\cdotp 10^{-11}[ m]\rightarrow (\mathrm{Bohr\ radius})\\
    c_{m} =m_{e} =9.109383\cdotp 10^{-31}[ kg]\rightarrow (\mathrm{Electron\ mass})\\
    c_{\hbar } =c_{m}( c_{x})^{2}( c_{t})^{-1} \Rightarrow c_{t} =\frac{c_{m}( c_{x})^{2}}{c_{\hbar }} =2.418884\cdotp 10^{-17}[ s] \equiv 0.02418884[ fs]
    \end{array}
\end{align}

entonces, si queremos evolucionar el sistema, por ejemplo hasta un tiempo real de $0.5[fs]$, tendremos que evolucionar durante un tiempo adimensional de;

\begin{align}
    t_{f} =0.5[ fs] =c_{t} \cdotp t_{f,ad} \Rightarrow t_{f,ad} =20.670689
\end{align}

Con esta elección de sitema de unidades tendremos que, los parámetros adimensionales del problema serán,

\begin{align}
    \begin{array}{l}
    \hbar _{ad} =1;\ \omega _{ad} =\frac{\omega }{\left( 2.418884\cdotp 10^{-17}[ s]\right)^{-1}} =1\Leftrightarrow \omega =\left( 2.418884\cdotp 10^{-17}[ s]\right)^{-1}\\
    \gamma _{ad} =\frac{\gamma }{\left( 9.109383\cdotp 10^{-31}[ kg]\right)\left( 5.291772\cdotp 10^{-11}[ m]\right)^{2}\left( 2.418884\cdotp 10^{-17}[ s]\right)^{-2}}\\
    \gamma _{ad} =\frac{\gamma }{\left( 4.359745\cdotp 10^{-18}[ J]\right)} =0.1\Rightarrow \gamma =4.359745\cdotp 10^{-19}[ J]
    \end{array}
\end{align}

entonces, la frecuencia de los Hamiltonianos tipo oscilador armónico tendrá un valor de uno numéricamente pero, corresponderá a frecuencias reales del orden de los exahertz,

\begin{align}
    \begin{array}{l}
    \omega =1\left[\frac{rad}{( au)_{t}}\right] =4.134137\cdotp 10^{16}\left[\frac{rad}{s}\right] =41.341378\left[\frac{rad}{fs}\right]\\
    \Rightarrow f=2\pi \omega =2.597554\cdotp 10^{17}[ Hz] =259,755544\left[( fs)^{-1}\right] \approx 2.6[ EHz]
    \end{array}
\end{align}

## Anexo II: Comparación entre FEM Julia vs MCTDH Package

### I. MCTDH Package

Para comparar los resultados obtenidos con el método de FEM en Julia se llevó a cabo una simulación utilizando un método numérico denominadoo [MCTDH](https://www.pci.uni-heidelberg.de/tc/usr/mctdh/lit/intro_MCTDH.pdf) por sus siglas en inglés Multiconfiguration time-dependent Hartree que es un algoritmo para propagar eficientemente paquetes de onda y por ello, es muy utilizado para estudiar dinámica cuántica. En particular, se utilizó el paquete de código abierto denominado MCTDH Heidelberg package que se ha estado desarrollando en la universidad de Heidelberg (Alemania) hace ya 18 años y permite tratar con grandes sistemas, es decir, sistemas con múltiples grados de libertad. Dicho paquete se encuentra desarrollado principalmente en los lenguajes de programación C y Fortran.

En la carpeta finla_project/MCTDH_simulations/03_version_components/ se encuentran los detalles de las configuraciones llevadas a cabo para el uso de MCTDH. A modo de comentario, podemos decir que se utilizó una ejecución en paralelo utilizando 4 threads (específicamente con método de paralelismo POSIX) y los siguientes parámetros de simulación:

+ Tiempo final de evolución de $100$
+ Intervalo temporal para guardar datos de $0.1$
+ Intervalo de tiempo para guardar función de onda de $0.1$
+ Base Sine-DVR como single particle function para discretización espacial, utilizando $501$ puntos de grilla en una caja de longitud $50$ simétrica respecto del origen.
Precisión de integradores numéricos del orden de $\approx \mathcal{O}\left( 10^{-6}\right)$.
